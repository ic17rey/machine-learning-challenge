```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(10)
```

```python
# View the columns from the data
df.columns
```

# Select your features (columns)

```python
# Set features as X (the data minus the column to be used as y)
#selected_features = df.drop(columns='koi_disposition')
X = df.drop(columns='koi_disposition')
X
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y=df.koi_disposition
y
```

```python
# View how the breakout looks for y in the data
y.value_counts()
```

```python
# Import/use train_test_split 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_train.head(20)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Look at a row of the results to see it as an array
X_train_scaled[0]
```

```python
# Another row
X_train_scaled[1]
```

# Train the Model



```python
# Using LogisticRegression for the first model
# LogisticRegression() gave warning about max_iter, increased to 1000
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
# Score the training and test data
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
# Use shift tab to explore more of the options taht can be updated here
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[1,5,10,15]}
grid = GridSearchCV(model,param_grid,verbose=3)

# Tried also with penalty l1, l2, none. l2 is best and is also the default.
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
#print(grid.best_params_)
print('Best Penalty:', grid.best_estimator_.get_params()['penalty'])
print('Best C:', grid.best_estimator_.get_params()['C'])
print(grid.best_score_)
```

# Save the Model

```python
# save model to turn in to BCS
# Looking at other models next to determine if Logistic Regession is best model
import joblib
filename = 'logistic_regression.sav'
joblib.dump(grid, filename)
```
